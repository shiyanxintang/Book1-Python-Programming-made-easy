```python
#抛1000次质地均匀的骰子并可视化均值
```

```python
import random
import statistics
import matplotlib.pyplot as plt
```

```python
# 抛骰子实验的次数
num_dices = 1000
# 用于存储每次抛骰子的结果
results = []  
# 用于存储每次抛骰子后的均值
running_means = []  
```

```python
random.seed(0)
for _ in range(num_dices):
    # 随机抛骰子
    # random.randint产生整数型随机数
    result_idx = random.randint(1,6)
    results.append(result_idx)

    # 计算当前所有结果的均值
    mean_idx = statistics.mean(results)
    running_means.append(mean_idx)
```

```python
# 可视化均值随次数变化
plt.plot(range(1, num_dices + 1), running_means)
plt.axhline(3.5, color = 'r')
plt.xlabel("Number of dices")
plt.ylabel("Running Mean")
plt.grid(True)
plt.ylim(0,6)
plt.show()
```
